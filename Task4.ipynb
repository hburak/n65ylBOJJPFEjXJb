```python
# bu sayfada economic_calendar veri setinin üzerinde Türkiye,Japonya ve Avustraly için analiz yapılmıştır.
import pandas as pd
econocal =pd.read_csv ("economic_calendar.csv")
df = econocal.copy()
```

```python
df=df[(df['date'] >= '2021-04-28') & (df['date'] < '2021-10-18')]
```

```python
df.tail ()
```

```python
# priority kısmı 3 olanların sayısını aldık:
print(df['priority'].value_counts().loc[lambda x : x>=3])
```

```python
# priority kısmı sadece türkiye için 2 ve üstü olanların sayısını aldık:
len(df [( df.country == "Türkiye") & (df.priority>=2 )])
```

```python
trdf =df [( df.country == "Türkiye") & (df.priority>=2 )]
```

```python
# değerleri açıklanmamış bir çok veri var, bunları silmemiz daha doğru. Çünkü olumlu ya da olumsuz etkisini bilmemiz gerekir.
trdf =trdf[trdf.explained.notnull()]
```

```python
#elimizde değerleri explained kısmında açıklanmış 16 veri kaldı
len (trdf)
```

```python
trdf
```

```python
#diğer ülkeler için girilen verilere baktık
print(df['country'].value_counts().loc[lambda x : x>20])
```

```python
#diğer ülkeler için açıklaması explained kısmında girilmemiş verileri filtereleyerek yoksaydık:
priority_df =df[df.priority>=2]
df2 =priority_df.explained.notnull().groupby(df['country']).sum().astype(int).reset_index(name='count')
print(df2.head(20))
```

```python
# bunları en yüksekten en düşüğe sıraladık
df2.sort_values(by=['count'], ascending=False).head(20)
```

```python
# türkiye için pivot table ile analiz 
trdf=econocal.copy()
trdf =trdf [( trdf.country == "Türkiye") & (trdf.priority>=2 )]
trdf =trdf[trdf.explained.notnull()]
len (trdf)
trdf
```

```python
trdf['year'] = pd.DatetimeIndex(trdf['date']).year
trdf['month'] = pd.DatetimeIndex(trdf['date']).month
```

```python
import numpy as np
pd.pivot_table(trdf, values='pk', index=['year'], columns=["month"], aggfunc=np.size)
```

```python
# avustralya için pivot table ile analiz 
avdf=econocal.copy()
avdf =avdf [( avdf.country == "Avustralya") & (avdf.priority>=2 )]
avdf =avdf[avdf.explained.notnull()]
len (avdf)
```
